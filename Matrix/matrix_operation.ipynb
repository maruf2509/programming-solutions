Task1:Display original & transpose

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of col: "))

print("Enter the matrix elements row wise:")
matrix = []
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    matrix.append(row)
```

```python
print("Original Matrix:")
for row in matrix:
    print(row)

transpose = [[matrix[j][i] for j in range(rows)] for i in range(cols)]

print("Transposed Matrix:")
for row in transpose:
    print(row)
```

Task2: Symmetric Matrix Check

```python
n = int(input("Enter size of square matrix "))

print("Enter the matrix elements row wise:")
matrix = []
for i in range(n):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    matrix.append(row)
```

```python
# symmetric Rule A == A^T
is_symmetric = all(matrix[i][j] == matrix[j][i] for i in range(n) for j in range(n))

print("\nMatrix:")
for row in matrix:
    print(row)

if is_symmetric:
    print("Symmetric.")
else:
    print(" NOT Symmetric.")
```

Task3:Skew-Symmetric Matrix Check

```python
n = int(input("Enter size of square matrix "))

print("Enter the matrix elements row by row:")
matrix = []
for i in range(n):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    matrix.append(row)
```

```python

is_skew = all(matrix[i][j] == -matrix[j][i] for i in range(n) for j in range(n))

print("\nMatrix:")
for row in matrix:
    print(row)

if is_skew:
    print(" The matrix IS Skew-Symmetric.")
else:
    print(" The matrix is NOT Skew-Symmetric.")
```
